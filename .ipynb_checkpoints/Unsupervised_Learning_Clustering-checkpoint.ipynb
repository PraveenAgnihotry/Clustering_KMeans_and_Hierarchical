```python
# Importing the initially required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
os.environ["OMP_NUM_THREADS"] = '2' # Just to remove some few warnings
```

```python
# Reading the .csv dataset file as a dataframe 
raw_df = pd.read_csv(r'drilling.csv')
df = raw_df
```

```python
# EDA
df.head()
```

```python
df.describe()
```

```python
# Scatter plot of the dataset
plt.scatter(df['x'], df['y'])
plt.show()
```

<h3> To determine the optimal number of clusters

```python
from sklearn.cluster import KMeans

inertias = []
# Testing the inertia with of upto 10 clusters
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[['x', 'y']])
    inertias.append(kmeans.inertia_)
```

```python
# Plotting the elbow curve of No of clusters vs Inertia
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow curve')
plt.xlabel('No. of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# We can notice a sharpe bend at no of clusters = 4, and after that there are no drastic changes in inertia.
# Hence we choose the number of clusters = 4
```

<h3> K-means clustering

```python
# Fitting the K-means model with 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(df[['x', 'y']])
# Saving the respective cluster values into the dataframe
df['kmeans_cluster'] = kmeans.fit_predict(df[['x', 'y']])
```

```python
# Performance metrics - Inertia and Silhouette Score
from sklearn.metrics import silhouette_score
inertia_km = kmeans.inertia_
silhouette_km = silhouette_score(df[['x', 'y']], df['kmeans_cluster'])
```

```python
# Visualization of K-means clusters
plt.scatter(df['x'], df['y'], c=df['kmeans_cluster'])
plt.title("K-Means Clustering")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<h3> Hierarchical clustering

```python
# Fitting the Hierarchical clustering model with 4 clusters
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

hierarchical = AgglomerativeClustering(n_clusters=4, linkage='ward')
# Saving the respective cluster values into the dataframe
df['hierarchical_cluster'] = hierarchical.fit_predict(df[['x', 'y']])
# Performance metric - Silhouette Score
silhouette_hc = silhouette_score(df[['x', 'y']], df['kmeans_cluster'])
```

```python
# Visualization of Hierarchical clusters
linked = linkage(df[['x', 'y']], method='ward')
dendrogram(linked, truncate_mode='lastp', p=10)
plt.title("Hierarchical Clustering")
plt.xlabel("Index")
plt.ylabel("Distance")
```
